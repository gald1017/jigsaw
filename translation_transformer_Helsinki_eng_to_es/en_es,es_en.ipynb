**Translate espaniol to english**

```python
from transformers import MarianMTModel, MarianTokenizer
model_name = 'Helsinki-NLP/opus-mt-es-en'
# Download the model and the tokenizer
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
                
```

```python
def translate(model,tokenizer,text):
      with torch.no_grad():
        text = text[0:509]
        model.to('cuda')
        tokens = tokenizer(text, return_tensors="pt", padding=True).to('cuda')
        tokens["input_ids"] = tokens["input_ids"][:,:512]
        tokens["attention_mask"] = tokens["attention_mask"][:, :512]
        translated = model.generate(**tokens).to('cuda')
        return [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
```

```python
type(tokenizer)
```

```python
import pickle
import itertools

texts = [es_text]
files_name = ['validation_es_to_en_Translation_Helsinki.pkl']


for i,t in enumerate(texts):
    translation = []
    start = time.time()
    for com in t:
        translation.append(translate(model,tokenizer,com))
        print(time.time())
    with open(files_name[i], 'wb') as f:
        torch.save(list(itertools.chain(*translation)), f)    
    end = time.time()
    print(end - start)
```
